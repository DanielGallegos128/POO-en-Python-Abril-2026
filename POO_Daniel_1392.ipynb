#### Conceptos de POO (Programación Orientada a Objetos) en Python

#### El prompt de la IA
# Lenguaje Python, crear la clase animal con los atributos "Nombre" y "Raza", crear el contructor animal, y las funciones ladrar(), correr(), crear los objetos Pitbull, Chihuahua, nivel principiante no es interactivo, proporciona tu los datos

```python
# Definición de la clase
class Animal:
    # El constructor (__init__) inicializa los atributos al crear el objeto
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    # Función (método) para ladrar
    def ladrar(self):
        print(f"{self.nombre} dice: ¡Guau guau! 🐶")

    # Función (método) para correr
    def correr(self):
        print(f"{self.nombre} está corriendo velozmente. 🏃‍♂️")

    def morder(self):
        print(f"{self.nombre} está mordiendo a la vecina. 🙎‍♀️")

# --- Creación de objetos con datos proporcionados ---

# Objeto 1: Pitbull
pitbull = Animal("Princesa", "Pitbull")

# Objeto 2: Chihuahua
chihuahua = Animal("Destructor de Galaxias", "Chihuahua")

# Creamos el objeto perro callejero
perro_callejero = Animal("Solovino", "Caramelo")

# --- Uso de los objetos ---

print(f"Perro 1: {pitbull.nombre}, Raza: {pitbull.raza}")
pitbull.ladrar()
pitbull.correr()

print("-" * 20)

print(f"Perro 2: {chihuahua.nombre}, Raza: {chihuahua.raza}")
chihuahua.ladrar()
chihuahua.correr()

print("-" * 20)

print(f"Perro 3: {perro_callejero.nombre}, Raza: {perro_callejero.raza}")
perro_callejero.ladrar()
perro_callejero.correr()
perro_callejero.morder()

print("Programa creado por Daniel Gallegos NL = 20")
```
